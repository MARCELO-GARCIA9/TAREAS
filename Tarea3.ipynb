```python
import numpy as np
import pandas as pd
```

```python
fallecidos=pd.read_csv('https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto50/DefuncionesDEIS_confirmadosPorComuna_std.csv')
vc1= pd.read_csv('https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto80/vacunacion_comuna_1eraDosis_std.csv')
riq= pd.read_csv('https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/hdi.csv')
```

```python
fallecidos=fallecidos[fallecidos.Region=='Metropolitana']
fallecidos=fallecidos[fallecidos.Comuna!='Total']
fallecidos['Fecha'] = pd.to_datetime(fallecidos['Fecha'], dayfirst=True)
fallecidos=fallecidos.sort_values(['Fecha'])
fallecidos=fallecidos.reset_index()
fallecidos=fallecidos.drop(['index'], axis=1 )
fallecidos=fallecidos.drop(['Codigo region'], axis=1 )
```

```python
fallecidos=fallecidos.drop(range(0,14999), axis=0)
fallecidos=fallecidos.reset_index()
fallecidos=fallecidos.drop(['index'], axis=1 )
fallecidos
```

```python
fallecidos=fallecidos.drop([10387], axis=0 )
fallecidos
```

```python
fallecidos2=fallecidos.groupby(['Comuna','Codigo comuna']).sum().reset_index()
fallecidos2=fallecidos2.drop(['Poblacion'], axis=1 )
fallecidos2
```

```python
vc1['Fecha']=pd.to_datetime(vc1['Fecha'], dayfirst=True)
vc1=vc1[vc1.Region=='Metropolitana']
vc1=vc1.reset_index()
vc1=vc1.drop(['index'], axis=1 )
vc1=vc1.drop(['Codigo region'], axis=1 )
vc1=vc1.drop([10599], axis=0 )
vc1=vc1.rename(columns={'Primera Dosis':'Primera'})
vc1
```

```python
vc1=vc1.groupby(['Comuna','Codigo comuna']).sum().reset_index()
vc1
```

```python
df_def=pd.merge(fallecidos2, vc1, on=['Comuna','Codigo comuna'], how='left')
```

```python
riq=riq.rename(columns={'comuna':'Codigo comuna'})
riq=riq.loc[:,['Codigo comuna','HDI']]
```

```python
df_def=pd.merge(df_def, riq, on=['Codigo comuna'], how='left')
```

```python
df_def
```

```python
import statsmodels.formula.api as sfm
```

```python
model=sfm.ols(formula='Defunciones ~ HDI+Primera',data=df_def)
fit = model.fit()
```

```python
fit.params
```

```python
print(fit.summary())
```

De acuerdo al R2, el 78.9% de la variacion de las defunciones por covid son explicadas por las variables de la cantidad de primeras dosis aplicadas y el HDI por comuna

El intercept, indica que si el HDI y el numero de primeras dosis son cero, las muertes serian de 470 personas

En cuanto a los factores podemos observar que a mayor riqueza(HDI) menor son las muertes y que de manera extraña aunque con baja influencia a mayor cantidad de primeras dosis mayor es la cantidad de muertes

Cabe notar que los resultados nos arrojan que puede existir una fuerte multicolinealidad entre las variables HDI y N° de primeras dosis 

Una variable que se podria agregar al modelo es la cantidad de segundas dosis aplicadas ya que de manera intuitiva la segunda dosis ofrece mayor proteccion y deberia hacer que el modelo se acerque un poco mas a la realidad actual que se esta viviendo.
