```python
import pandas as pd
```

```python
vc1= pd.read_csv('https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto81/vacunacion_comuna_edad_1eraDosis.csv')
vc2= pd.read_csv('https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto81/vacunacion_comuna_edad_2daDosis.csv')
riq= pd.read_csv('https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/hdi.csv')
pob= pd.read_csv('https://raw.githubusercontent.com/leoferres/gitiele356b/main/data/pobXedadCenso.csv')
```

```python
vc1_metro=vc1[vc1.Region=='Metropolitana']
vc2_metro=vc2[vc2.Region=='Metropolitana']
```

```python
vc1_metro=vc1_metro.reset_index()
vc1_metro=vc1_metro.drop(['index'], axis=1 )
vc1_metro=vc1_metro.drop([52], axis=0 )
vc1_metro['1era_EDAD_15A64'] = vc1_metro.iloc[:,5:55].sum(axis=1)
vc1_metro['1era_EDAD_65YMAS'] = vc1_metro.iloc[:,55:71].sum(axis=1)
vc1_metro['TOTAL_1era'] = vc1_metro.iloc[:,5:71].sum(axis=1)
vc1_metro
```

```python
vc1_metro2=vc1_metro.loc[:,['Region','Comuna','Codigo comuna','Poblacion','1era_EDAD_15A64','1era_EDAD_65YMAS','TOTAL_1era']]
```

```python
vc2_metro=vc2_metro.reset_index()
vc2_metro=vc2_metro.drop(['index'], axis=1 )
vc2_metro=vc2_metro=vc2_metro.drop([52], axis=0 )
vc2_metro['2da_EDAD_15A64'] = vc2_metro.iloc[:,5:55].sum(axis=1)
vc2_metro['2da_EDAD_65YMAS'] = vc2_metro.iloc[:,55:71].sum(axis=1)
vc2_metro['TOTAL_2da'] = vc2_metro.iloc[:,5:71].sum(axis=1)
vc2_metro
```

```python
vc2_metro2=vc2_metro.loc[:,['Region','Comuna','Codigo comuna','Poblacion','2da_EDAD_15A64','2da_EDAD_65YMAS','TOTAL_2da']]
```

```python
vc_metro=pd.merge(vc1_metro2, vc2_metro2, on=['Region','Comuna','Codigo comuna','Poblacion'], how='left')
```

```python
vc_metro
```

```python
riq=riq.rename(columns={'comuna':'Codigo comuna'})
riq=riq.loc[:,['Codigo comuna','HDI']]
```

```python
riq
```

```python
vc_metro=pd.merge(vc_metro, riq, on=['Codigo comuna'], how='left')
```

```python
pob['EDAD_0A5'] = pd.to_numeric(pob['EDAD_0A5'])
pob['EDAD_6A14'] = pd.to_numeric(pob['EDAD_6A14'])
pob['EDAD_15A64'] = pd.to_numeric(pob['EDAD_15A64'])
pob['EDAD_65YMAS'] = pd.to_numeric(pob['EDAD_65YMAS'])
```

```python
pob.dtypes
```

```python
pob=pob.groupby(['Codigo comuna']).sum()
```

Mientras se exploraban los datos me di cuenta que no coincidian la poblacion de cada comuna entregada por el dataset de vacunacion y la poblacion entregada por el dataset de la poblacion, dado que si se trabaja los datos de vacunacion con la poblacion entregada por el dataset de la poblacion existirian mÃ¡s vacunados que poblacion en algunas comunas. Por lo que se trabajara solo con la poblacion entregada por el dataset de la vacunacion

```python
vc_metro['%1eraVACUNA'] = vc_metro['TOTAL_1era']/vc_metro['Poblacion']*100
vc_metro['%2daVACUNA'] = vc_metro['TOTAL_2da']/vc_metro['Poblacion']*100
vc_metro['1era_%15A64'] = vc_metro['1era_EDAD_15A64']/vc_metro['Poblacion']*100
vc_metro['1era_%65YMAS'] = vc_metro['1era_EDAD_65YMAS']/vc_metro['Poblacion']*100
vc_metro['2da_%15A64'] = vc_metro['2da_EDAD_15A64']/vc_metro['Poblacion']*100
vc_metro['2da_%65YMAS'] = vc_metro['2da_EDAD_65YMAS']/vc_metro['Poblacion']*100
```

```python
vc_metro
```

```python
vc_metro.corr()
```

```python
import seaborn as sb

sb.lmplot(x = "HDI",
            y = "%1eraVACUNA",
            ci = None,
            data = vc_metro)
sb.lmplot(x = "HDI",
            y = "%2daVACUNA", 
            ci = None,
            data = vc_metro)
```

En estos graficos se compara el estrato socioeconomico de las diferentes comunas con su porcentaje de vacunados totales separados en los que tienen la primera y la segunda dosis, donde se puede observar que si bien los datos estan dispersos estos presentan una pendiente positiva por lo que si existe algun tipo de relacion de que las comunas mas ricas tienen mayor porcentaje de vacunados con la primera dosis. En el caso del porcentaje de vacunados con la segunda dosis comparados con el nivel socioeconomico, la pendiente es aun mayor por lo que la relacion es aun mayor.
Lo anteriormente dicho se puede observar al ver el coeficiente de correlacion entre la el %1eravacuna y el hdi es de 0.353208, mientras que en el caso del %2davacuna y el hdi la correlacion es aun mayor, siendo 0.447187 

```python
sb.lmplot(x = "HDI",
            y = "1era_%15A64", 
            ci = None,
            data = vc_metro)
sb.lmplot(x = "HDI",
            y = "1era_%65YMAS", 
            ci = None,
            data = vc_metro)
```

En estos graficos se analizo el porcentaje de vacunados con la primera dosis por tramo de edad de entre 15A64 y 65YMAS, no se analisaron los menores de 15 aÃ±os ya que estos no se han vacunados.

Al ver estos graficos se puede que en general el porcentaje de vacunados entre 15 y 64 aÃ±os es mayor en magnitud por lo amplio del tramo en comparacion al tramo de 65 y mas.

En 1era_%15a64 Y 2DA_%65YMAS se esta calculando el porcentaje de la poblacion total entre estos tramos de edad que tiene la primera o segunda dosis segun corresponda.

En los graficos mostrados se puede ver que la pendiente en el caso del porcentaje de la poblacion total con primera dosis entre los 15 y 64 aÃ±os, y el hdi es menor al caso del porcentaje de la poblacion total con primera dosis con 65 y mÃ¡s aÃ±os, y el hdi.
Tambien al comparar las correlaciones del primer caso (0.293718) es menor al caso de 65 aÃ±os y mas (0.403536) podemos ver a mayor edad, el porcentaje de vacunados tiene mayor relacion con el hdi.


```python
sb.lmplot(x = "HDI",
            y = "2da_%15A64", 
            ci = None,
            data = vc_metro)
sb.lmplot(x = "HDI",
            y = "2da_%65YMAS", 
            ci = None,
            data = vc_metro)
```

En los graficos mostrados se puede ver que la pendiente en el caso del porcentaje de la poblacion total con segunda dosis entre los 15 y 64 aÃ±os, y el hdi es menor al caso del porcentaje de la poblacion total con segunda dosis con 65 y mÃ¡s aÃ±os, y el hdi.
