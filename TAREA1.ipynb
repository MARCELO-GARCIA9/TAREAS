```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
from time import gmtime, strftime
```

```python
defunciones= pd.read_csv('BaseCorregida.csv',delimiter=';')
```

```python
df=defunciones
```

```python
df.shape
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['FECHA_DEF'] = pd.to_datetime(df['FECHA_DEF'], dayfirst=True)
```

```python
df.dtypes
```

```python
df
```

```python
df= df.rename(columns={'ANO_DEF':'AÑO',
                        'FECHA_DEF':'FECHA_DEFUNCION',
                        'GLOSA_SEXO':'SEXO',
                        'EDAD_CANT':'EDAD',
                        'CODIGO_COMUNA_RESIDENCIA.1':'COMUNA',
                        'GLOSA_REG_RES':'REGION',
                        'CODIGO_GRUPO_DIAG1':'CODIGO_DIAG_GENERAL',
                        'GLOSA_GRUPO_DIAG1':'DIAGNOSTICO_GENERAL',
                        'GLOSA_CATEGORIA_DIAG1':'DIAGNOSTICO_CATEGORIA',
                        'GLOSA_SUBCATEGORIA_DIAG1':'DIAGNOSTICO_FINAL'})
```

```python
df
```

```python
df1=df.drop(['EDAD_TIPO','CODIGO_COMUNA_RESIDENCIA','CAPITULO_DIAG1'], axis=1 )
```

```python
df1
```

```python
df1 = df1.sort_values(["FECHA_DEFUNCION"])
```

```python
df1
```

```python
df1=df1.reset_index()
df1=df1.drop(['index'], axis=1 )
df1
```

```python
grouper = pd.Grouper(key='FECHA_DEFUNCION', freq='W-MON')
df1["week"] = df1.groupby(grouper).grouper.group_info[0] + 1
```

```python
df1
```

```python
df_resumen_total=df1.groupby(['week']).size().reset_index(name='Defunciones')
```

```python
df_resumen_total
```

```python
plt.plot(df_resumen_total['week'],df_resumen_total['Defunciones'])
plt.xlabel('Semana')
plt.ylabel('Fallecidos')
plt.title('Fallecidos totales por semana desde 2016')
plt.show()
```

```python
df_gripe =df1.drop(df1[df1.CODIGO_CATEGORIA_DIAG1!='J09-J18'].index)
```

```python
df_gripe
```

```python
df_gripe=df_gripe.reset_index()
df_gripe=df_gripe.drop(['index'], axis=1 )
df_gripe
```

```python
df_resumen_gripe=df_gripe.groupby(['DIAGNOSTICO_CATEGORIA','week']).size().reset_index(name='Defunciones')
```

```python
df_resumen_gripe
```

```python
plt.plot(df_resumen_gripe['week'],df_resumen_gripe['Defunciones'],label='GRIPE')
plt.xlabel('N° de Semana')
plt.ylabel('Fallecidos')
plt.title('Fallecidos por Gripe por semana desde 2016')
plt.show()
```

```python
plt.plot(df_resumen_gripe['week'],df_resumen_gripe['Defunciones'],label='GRIPE')
plt.plot(df_resumen_total['week'],df_resumen_total['Defunciones'],label='TOTAL')
plt.xlabel('N° de Semana')
plt.ylabel('Fallecidos')
plt.legend(loc="upper left")
plt.title('Fallecidos de gripe y totales por semana desde 2016')
plt.show()
```

```python
df_covid =df1.drop(df1[df1.CODIGO_CATEGORIA_DIAG1!='U00-U49'].index)
```

```python
df_covid
```

```python
df_covid=df_covid.reset_index()
df_covid=df_covid.drop(['index'], axis=1 )
df_covid
```

```python
df_covid['GLOSA_CAPITULO_DIAG2']=df_covid['GLOSA_CAPITULO_DIAG2'].replace(['COVID-19 virus no identificado', 'COVID-19 virus identificado'], 'COVID19')
```

```python
df_covid
```

```python
df_resumen_covid=df_covid.groupby(['GLOSA_CAPITULO_DIAG2','week']).size().reset_index(name='Defunciones')
```

```python
df_resumen_covid
```

```python
plt.plot(df_resumen_covid['week'],df_resumen_covid['Defunciones'],label='COVID19')
plt.legend(loc="upper left") 
plt.xlabel('Semana')
plt.ylabel('Fallecidos')
plt.title('Fallecidos por COVID por semana desde 2020')
plt.show()
```

```python
plt.plot(df_resumen_covid['week'],df_resumen_covid['Defunciones'],label='COVID19')
plt.plot(df_resumen_total['week'],df_resumen_total['Defunciones'],label='TOTAL')
plt.legend(loc="upper left") 
plt.xlabel('Semana')
plt.ylabel('Fallecidos')
plt.xlim(200,278)
plt.title('Fallecidos por COVID vs Fallecidos Totales por semana desde 2016')
plt.show()
```

```python
plt.plot(df_resumen_covid['week'],df_resumen_covid['Defunciones'],label='COVID19')
plt.plot(df_resumen_total['week'],df_resumen_total['Defunciones'],label='TOTAL')
plt.plot(df_resumen_gripe['week'],df_resumen_gripe['Defunciones'],label='GRIPE')
plt.legend(loc="upper left") 
plt.xlabel('Semana')
plt.ylabel('Fallecidos')
plt.title('Fallecidos por COVID vs Fallecidos Totales por semana desde 2016')
plt.show()
```

Como se puede ver en este grafico que agrupa las muertes totales, por gripe, y por covid19, las muertes por gripe se han mantenido constante a lo largo de las semanas hasta el dia de hoy, mientras que cuando aparecio el covid se ve una clara disrupcion en el ritmo de muertes totales por semana.
Por otro lado podemos ver que el nivel de muertes causada por covid es muy superior a las causadas por la gripe, por lo que no se podria decir que el covid es como la gripe
